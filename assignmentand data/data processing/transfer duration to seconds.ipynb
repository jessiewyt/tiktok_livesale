```python
def duration_chinese_to_seconds(x):
    if not x is np.nan:
        print(x)
        seconds = 0
        if '小时' in x:
            hr = int(x[:x.index('小时')])
            #print(hr)
            seconds += hr*60*60
            x = x[x.index('小时')+2:]
            #print(x)
        if "分钟" in x:
            mins = int(x[:x.index("分钟")])
            #print(mins)
            seconds += mins*60
            x = x[x.index("分钟")+2:]
            #print(x)
        if "秒" in x:
            seconds += int(x[:x.index("秒")])
        return seconds
```

```python
for i in dt.Duration.tolist():
    print(duration_chinese_to_seconds(i))
```

```python
dt['seconds'] = dt['Duration'].apply(duration_chinese_to_seconds)
```
